# 5. 베스킨라빈스, 던킨도너츠 이름 정의

```python
# 베스킨라빈스, 던킨도너츠는 다양하게 상호명이 존재함. 하나로 통일하기.
df_seoul["상호명_정의"] = df_seoul["상호명"].str.lower()
```

```python
df_seoul[df_seoul["상호명_정의"].str.contains("던킨|dunkin")]
```

```python
df_31 = df_seoul[df_seoul["상호명_정의"].str.contains("던킨|dunkin|(베|배)스킨라빈스|baskinrobbins")].copy()
```

```python
df_31.loc[df_31["상호명_정의"].str.contains("배스킨라빈스|베스킨라빈스|baskinrobbins"),
          "브랜드명"] = "배스킨라빈스"
```

```python
df_31["브랜드명"] = df_31["브랜드명"].fillna("던킨도너츠")
```

```python
df_31["상권업종대분류명"].value_counts()
```

```python
df_31[df_31["상권업종대분류명"].isin(["소매", "생활서비스"])]
```

```python
df_31 = df_31[-df_31["상권업종대분류명"].isin(["소매", "생활서비스"])]
```

# 6. 배스킨라빈스, 던킨도너츠 지도에 나타내기

```python
import folium
```

```python
lat = df_31["위도"].mean()
```

```python
long = df_31["경도"].mean()
```

```python
m = folium.Map([lat, long])
```

```python
df_31.index
```

```python
for i in df_31.index :
    sub_lat = df_31.loc[i, "위도"]
    sub_long = df_31.loc[i, "경도"]
    sub_name = df_31.loc[i, "상호명"]+" - "+df_31.loc[i, "도로명주소"]
    
    icon_color = "blue"
    if df_31.loc[i, "브랜드명"] == "던킨도너츠" :
        icon_color = "red"
        
    folium.Marker([sub_lat, sub_long],
                 icon = folium.Icon(color=icon_color),
                 toolip = sub_name).add_to(m)
    
```
