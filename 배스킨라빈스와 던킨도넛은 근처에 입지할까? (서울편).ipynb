## 4. 배스킨라빈스, 던킨도너츠 이름 정의하기 
### (다양한 종류의 이름이 존재함. 이름하나로 정의하기)

```python
#1. 이름 소문자로 만들기
df["상호명_정의"] = df["상호명"].str.lower() ]
```

```python
#2. 상호명_정의 배스킨라빈스 찾기
df["상호명_정의"].str.extract("(배|베)스킨라빈스|baskinrobbins").value_counts()
```

```python
df[df["상호명_정의"].str.contains("(배|베)스킨라빈스|baskinrobbins")]
```

```python
#3. 상호명_정의 던킨도넛 찾기
df[df["상호명_정의"].str.contains("던킨|dunkin")]
```

```python
#4. 배스킨라빈스 이름 정의
df_31 = df[df["상호명_정의"].str.contains("(배|베)스킨라빈스|baskinrobbins|던킨|dunkin")].copy()
df_31
```

```python
df_31.loc[df_31["상호명_정의"].str.contains("(배|베)스킨라빈스|baskinrobbins"), "브랜드명"] = "배스킨라빈스"
```

```python
#4. 던킨도너츠 이름 정의
df_31["브랜드명"] = df_31["브랜드명"].fillna("던킨도너츠")
```

```python
#5. 상권업종대분류명 != 음식 항목 제외하기
df_31["상권업종대분류명"].value_counts()
```

```python
df_31[df_31["상권업종대분류명"].isin(["소매"])]
```

```python
df_31 = df_31[-df_31["상권업종대분류명"].isin(["소매"])]
```

```python
df_31["상권업종대분류명"].value_counts()
```

## 5. 지도에 나타내기

```python
import folium
```

```python
lat = df_31["위도"].mean()
```

```python
long = df_31["경도"].mean()
```

```python
m = folium.Map([lat, long])
```

```python
df_31.index
```

```python
for i in df_31.index :
    sub_lat = df_31.loc[i, "위도"]
    sub_long = df_31.loc[i, "경도"]
    sub_title = df_31.loc[i, "상호명"] + " - " + df_31.loc[i, "도로명주소"]
    
    color = "pink"
    if df_31.loc[i, "브랜드명"] == "던킨도너츠":
        color="orange"
        
    folium.Marker([sub_lat, sub_long],
                 icon = folium.Icon(color=color),
                 toolip = sub_title).add_to(m)
    
m
```
